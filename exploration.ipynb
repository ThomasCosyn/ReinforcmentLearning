## II. Encodage

```python
trigo_encoding = ["Month", "Day Type", "Hour"]
min_max_encoding = []
df_enc = df.copy()
for col in names:
    if col in trigo_encoding:
        df_enc[col + "_cos"] = np.cos((df_enc[col] * 2 * np.pi)/stats['max'][[col]][0])
        df_enc[col + "_sin"] = np.sin((df_enc[col] * 2 * np.pi)/stats['max'][[col]][0])
    else:
        df_enc[col] = df_enc[col]/(stats['max'][[col]][0] - stats['min'][[col]][0])

df_enc
```

## III. Linear regression and feature importance

```python
df
```

```python
trigo_encoding = ["Month", "Day Type", "Hour"]
df_LR = df.copy()
for col in names:
    if col in trigo_encoding:
        df_LR[col + "_cos"] = np.cos((df_LR[col] * 2 * np.pi)/stats['max'][[col]][0])
        df_LR[col + "_sin"] = np.sin((df_LR[col] * 2 * np.pi)/stats['max'][[col]][0])

df_LR["C"] = df_LR["Price"] * df_LR["Power consumption"]
df_LR["G"] = df_LR["Carbon density"] * df_LR["Power consumption"]
```

```python
df_LR.columns
```

```python
cols = ["Temperature", "Temperature + 6h", "Temperature + 12h", "Temperature + 24h", "Humidity rate", "Humidity rate + 6h", "Humidity rate + 12h", "Humidity rate + 24h", "Diffuse sun", "Diffuse sun + 6h", "Diffuse sun + 12h", "Diffuse sun + 24h", "Direct sun", "Direct sun + 6h", "Direct sun + 12h", "Direct sun + 24h", "Carbon density", "Equipment power", "Solar Power", "Battery storage", "Power consumption", "Price", "Price + 6h", "Price + 12h", "Price + 24h", "Month_cos", "Month_sin", "Day Type_cos", "Day Type_sin", "Hour_cos", "Hour_sin"]
X = df_LR[cols]
yC = df_LR[["C"]]
yG = df_LR[["G"]]
regC = LinearRegression().fit(X, yC)
regG = LinearRegression().fit(X, yG)
```

```python
regC.coef_
```

```python
fimportanceC = pd.DataFrame([[cols[i], abs(regC.coef_[0][i])] for i in range(len(cols))], columns = ['Variable', '|Coef|'])
fimportanceG = pd.DataFrame([[cols[i], abs(regG.coef_[0][i])] for i in range(len(cols))], columns = ['Variable', '|Coef|'])
```

```python
fimportanceC.astype({'|Coef|' : 'float64'}).dtypes
fimportanceC.sort_values(by = "|Coef|", ascending = False).head(10)
```
