## I. Exploration

```python
import pandas as pd
import numpy as np
```

```python
names = ["Month", "Day Type", "Hour", "Temperature", "Temperature + 6h", "Temperature + 12h", "Temperature + 24h", "Humidity rate", "Humidity rate + 6h", "Humidity rate + 12h", "Humidity rate + 24h", "Diffuse sun", "Diffuse sun + 6h", "Diffuse sun + 12h", "Diffuse sun + 24h", "Direct sun", "Direct sun + 6h", "Direct sun + 12h", "Direct sun + 24h", "Carbon density", "Equipment power", "Solar Power", "Battery storage", "Power consumption", "Cost", "Cost + 6h", "Cost + 12h", "Cost + 24h"]
df = pd.read_csv("observations.csv", delimiter = ";", names = names)
df
```

```python
stats = df.describe().transpose()
stats
```

## II. Encodage

```python
trigo_encoding = ["Month", "Day Type", "Hour"]
min_max_encoding = []
df_enc = df.copy()
for col in names:
    if col in trigo_encoding:
        df_enc[col + "_cos"] = np.cos((df_enc[col] * 2 * np.pi)/stats['max'][[col]][0])
        df_enc[col + "_sin"] = np.sin((df_enc[col] * 2 * np.pi)/stats['max'][[col]][0])
    else:
        df_enc[col] = df_enc[col]/(stats['max'][[col]][0] - stats['min'][[col]][0])

df_enc
```
